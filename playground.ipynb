```python
import pandas as pd

# Load the normalized transaction data (adjust path if needed)
df = pd.read_csv("output/transactions_normalized.csv", parse_dates=["timestamp"])

# Preview
df.head()
```

```python
# List of transaction types relevant to crypto (adjust if needed)
crypto_types = [
    "buy", "sell", "staking_reward", "airdrop",
    "transfer_in", "transfer_out", "deposit", "withdrawal", "swap"
]

# Get distinct asset symbols used in crypto-related transactions
crypto_assets = (
    df[df["type"].isin(crypto_types)]
    .asset.dropna()
    .str.upper()
    .unique()
)

# Display sorted list for mapping
sorted_assets = sorted(crypto_assets.tolist())
print(sorted_assets)
```
